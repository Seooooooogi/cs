# 1. 정보적 관점에서의 entropy

- 각 메시지에 포함된 정보의 기댓값(평균), 무질서도, 불확실성
- 확률이 낮을수록 어떤 정보일지 불확실하게 되고, 이때 정보가 많다(=엔트로피가 높다)라고 표현
- 사건이 불확실하면 그 사건이 일어났을 때 훨씬 유용한 정보를 제공하기 때문에 정보량은 확률에 반비례
- 발생 가능한 사건이나 메시지의 확률분포의 음의 로그로 정의

만약 E가 이산공간 $x_1,...x_n$이고 확률변수 $X:P \to E$가 분포 $f: E \to \mathbb{R} $을 따른다고 할 때 $X$의 **shannon 정보 엔트로피 $H(X)$** 는 다음과 같음


$$H(X) = - \sum_{i=1}^{n}p_{i}\log_{2}p_i$$

## ex) 동전 던지기

- 동전 던지기를 시행했을 때 결과가 공정하다면 엔트로피는 가장 높게 나옴
- 앞,뒷면이 동일하게 1/2 확률이라면 불확실성이 가장 커지고 예측하기 어렵기 때문
- 이때의 결과값은 1비트에 해당하는 정보를 가지게 됨.
$$-\frac{1}{2}\log_{2}\frac{1}{2} + -\frac{1}{2}\log_{2}\frac{1}{2} = 1$$
- 만약 동전 던지기가 공정하지 않다면 불확실성은 떨어지게 됨
- 이 경우에는 동전 던지기의 엔트로피는 1비트의 정보보다 적다고 해석할 수 있음
- 만약에 극단적인 경우(항상 앞면이 나옴)에는 엔트로피는 0이 됨. 즉 아무런 정보도 전달하지 못함 : 어떠한 경우에도 앞면이나 뒷면이 나올 것이 확실하기 때문
$$-1\log_{2}1  -0 \times \log_{2}0$$


## 엔트로피의 기본적인 성질

- 성질 1 : Uniform distribution은 불확실성이 가장 큼
- 성질 2 : 불확실성은 독립일 때 더할 수 있음

$$H(X,Y) = H(X) + H(Y)$$

- 두 개의 코인을 생각할 때, 첫번째 코인은 Head일 확률이 80%, tail일 확률이 20%이고 두번째 코인은 Head일 확률이 60%, tail일 확률이 40%라고 했을 때 두 코인을 동시에 던졌을 때의 엔트로피는
$$H(\{0.48,0.32,0.12,0.08\}) = H(\{0.8,0.2\}) + H(\{0.6,0.4\})$$


- 성질 3 : 0% 확률의 엔트로피 결과를 더하는 것은 의미 없음  
$$H(\{0.8,0.2\}) = H(\{0.8,0.2,0\})$$


- 성질 4 : 연속성 - 불확실성은 모든 독립 변수들에 연속적임

- 성질 5 : 더 많은 outcome이 있는 uniform distribution이 더 불확실성이 높음
- L이 uniform distibution이라고 할 때,

$$L(m) > L(n)\; for\; m > n$$

$$H(0.5,0.5) < H(1/6,1/6,1/6,1/6,1/6,1/6)$$

- 성질 6 : 엔트로피는 항상 0보다 크다
$$H(X) = - \sum_{i=1}^{n}p_{i}\log_{2}p_i$$
확률은 항상 0과 1사이의 값이기 때문에 $-\log_{2}p_i$는 항상 0보다 크다
***엔트로피가 0이 될 때 : 그 사건이 무조건 일어나는 경우**


# 2. cross-entropy

- cross entropy는 두 확률분포를 비교해서 얼마나 다른지 측정
- cross entropy를 Q(x)를 사용할 때 P(x)로부터 그려지는 이벤트의 정보량의 기대량

$$H(P,Q) = - \sum_{x}^{}P(x)\log \{X\}_{2}Q(x)$$

- 어떤 경우가 cross entropy가 entropy보다 높을까?


- 왼쪽의 경우 cross entropy는 entropy와 동일  
왜? Q(x)의 정보가 Q(x)와 유사한 P(x)의 분포에 따라 가중치가 부여되기 때문  
- 반면 오른쪽의 경우에 P(x)와 Q(x)는 다르기 때문에 cross entropy가 더 높음  
왜? 많은 양의 정보가 가중치가 적은 반면에 적은 양의 정보가 가중치가 크기 때문  
즉, Q(x)의 불확실성이 클 때 P(x)의 불확실성이 작다는 것은 두 분포가 유사하지 않다는 것을 의미

- cross entropy는 두 분포가 동일할 때 최소
- cross entropy는 엔트로피보다 낮을 수 없음. : Q(x)가 P(x)보다 크다면 반대편에서는 P(x)가 Q(x)보다 커서 균형을 잡아주기 때문

- 또한, $H(P,Q) \neq H(Q,P)$

- cross entropy는 실제 분포와 모델에 의해 추정되는 분포를 비교할 때 사용됨

- 이 모델은 소리로 새의 종류를 3가지로 분류하는 모델

이때의 cross entropy는 다음과 같음.
$$
\begin{align}
H(P,Q) &= - \sum_{x}^{}P(x)\log Q(x)\\
& =-(1.0\log 0.7 +0.0\log 0.2+0.0\log 0.1)\\
& =-\log 0.7
\end{align}
$$

# 3. KL divergence
- KL divergence는 두 분포 $P(x)$와 $Q(x)$의 엔트로피 차이 

$$\begin{align}
D_{KL}(P||Q) &= H(P,Q) - H(P)\\
&= - \sum_{x}^{}P(x)\log Q(x) - (- \sum_{x}^{}P(x)\log P(x))\\
&= - \sum_{x}^{}P(x)\log \frac{P(x)}{Q(x)}
\end{align}
$$
- KL divergence는 항상 0보다는 큼 : cross entropy는 entropy보다 낮을 수 없기 때문 , 즉 두 분포가 동일하다면 KL divergence는 0임

- KL divergence가 거리가 될 수 없는 이유 :  
$$
KL(P \parallel Q)=  \sum_xP(x)log\frac{P(x)}{Q(x)} \neq \sum_xQ(x)log\frac{Q(x)}{P(x)} = KL(Q \parallel P)
$$
- $KL(P \parallel Q)$와 $KL(Q \parallel P)$가 다르기 때문
- 이를 해결하기 위해서 JS divergence 고안 : symmetric해짐
$$\text{JSD}(p,q) = \frac{1}{2}D_ {KL} (p||\frac{p+q}{2}) + \frac{1}{2}D_ {KL} (q||\frac{p+q}{2})$$

### 왜 KL divergence를 쓰지 않고 Cross Entropy를 쓸까?
- 우리가 최적화하고 싶은 값은 $Q(x)$, 즉 $Q(x)$의 입장에서 보면 $P(x)$는 상수이기 때문에, CE에 대해서만 최적화해주면 되기 때문

# 4. curse of dimensionality

- 차원을 늘리는 것은 축복이 될 수도 있고 저주가 될 수도 있음

## 1) 축복인 경우
- 저차원에서 복잡했던 feature가 고차원에서는 선형 분류가 가능함  
  ex) kernel method : 저차원에서는 왼쪽 같은 원형 특징은 선형 문제로 풀 수가 없으나 커널을 사용하여 feature를 저차원으로 고차원으로 transform시킴


## 2) 저주인 경우
- 예를 들어, 각 차원마다 두 점 사이의 거리를 구해보자.
1. 1차원 : $D(A,B) = x2-x1$
2. 2차원 : $D(A,B)= \sqrt{(x2-x1)^2+(y2-y1)^2}$
3. 3차원 : D(A,B)~ sqrt(3)
4. n차원 : D(A,B)~ sqrt(4)

- 즉, 고차원일수록 두 점 사이의 거리에 공간이 넓어짐. 즉 전체적인 공간이 커짐으로 밀도를 유지하기 위해서 더 많은 점들을 샘플해야함. 즉 더 많은 데이터를 필요로 함

- 머신 러닝의 관점에서, optimizer가 고차원의 빈 공간에서 길을 잃고 잘 수렴하지 못할 수 있기 때문에 이것은 학습 시간이 늘어나고 최적의 답을 찾지 못할 수도 있음.
- 즉, 축복을 최대한 받고 저주를 최대한 피할 수 있는 최적의 차원을 찾는 것이 중요
